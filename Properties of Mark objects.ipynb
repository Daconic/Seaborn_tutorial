# Properties of Mark objects

## Coordinate properties

### x,y,xmin,xmax,ymin,ymax

- 좌표 특성은 그래프에서 어디에 마크할 지 결정합니다. 
- 표준에서 x 는 수평 포지션 그리고 y 는 수직 포지션을 담당합니다. 
- 어떤 마크는 하나 또는 두 개의 변수들의 폭(min, max)을 수용합니다. 
- `baseline` 파라미터로 폭을 보여줄 수도 있습니다
- 레이어의 `orient` 파라미터가 어떤 방식으로 작동하는지 결정합니다. 

- 만약 변수가 수치형 데이터를 가지고 있지 않다면 화면에 그려질 수 있도록 데이터는 변형될 것입니다. 
- 예를 들어 `Nominal`(순서형) 스케일은 각 카테고리에 정수 인덱스로 지정되고 `Temporal` 스케일들은 날짜형으로 나타납니다. 

- `Continuus`(연속형) 스케일들은 데이터 갑과의 공간적 위치로 비선형으로 변형됩니다. 

## Color properties

### color, fillcolor, edgecolor

- 모든 마크들은 `color`가 주어집니다 그리고 마크의 `edge` 또는 `fill`로 구분합니다. 
- 가끔, 단순이 `color`로 더 많은 조정을 할 수 있습니다.

- 색깔의 특성이 정해지면 기본 팔레트가 스케일의 종류에 의존합니다. 
- 순서 스케일은 discrete을 사용하고 순서가 없으면 hue 반면에 연속형 스케일은 연속적인 변화도를 사용합니다.

- 색깔 스케일은 `viridis`, `rocket`, or `deep`과 같은 팔레트의 이름으로 파라미터화할 수 있습니다. 
- 어떤 팔레트에 간단한 변화를 포함 파라미터를 가지고 있습니다.(e.g. `dark:blue`) 또는 cubehelix 시스템(e.g. `ch:start=.2,rot=-.4`)

- 연속형 스케일은 색깔의 튜플로 파라미터화 할 수 있습니다.
- 순서형 스케일을 사용합니다 팔레트의 이름, 각 색깔의 리스트, 또는 색깔의 딕션너리 형태로 지정할 수 있습니다. 

- 색깔들은 다양한 범위의 포멧으로 특정될 수 있습니다. 
- 이 중에는 색깔 사이클(`C0`), 한 글자 단축키(`b`), 그레이 스케일 값(`.4`), RGB 헥스 코드(`#472b0`), X11 색 이름(`seagreen`), 그리고 XKCD 색 조사 이름(`purpleish`)
