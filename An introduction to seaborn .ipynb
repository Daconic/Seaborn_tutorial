- 따라서, matplotlib를 어떤식으로 사용할 지에 대해 고민하는것 보다 던져진 질문에 답을 하는것에 집중할 수 있게 도와줍니다.

## A high-level API for statistical graphics

- 데이터를 시각화 하는데 가장 좋은 방법은 없습니다! 
- 서로 다른 질문들의 가장 좋은 답은 서로 다른 그래프로 보여집니다. 
- Seaborn은 지속 가능한 데이터 기반 API을 통해 다양한 시각화 표현을 쉽게 전환할 수 있게 해줍니다. 
- replot() 함수는 서로 다른 통계학적 관계를 시각화하기 위해서 디자인되었다. 
- Scatter plot은 하나의 효과적인 시각화 방법이지만 하나의 변수의 시간을 측정한 관계는 선으로 표현하는 것이 낫다. 
- replot() 함수는 kind라는 파라미터로 그래프를 편하게 변경할 수 있다. 

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False))
```

- size 와 style 파라미터는 scatter plot과 line plot에 모두 사용되었다. 하지만 막상 보면 시각화 시에 다른 영향을 끼쳤다. 
- scatter plot의 경우는 마커부분과 기호를 바꿨고 line plot에서는 선의 굵기와 대시선으로 바꿨습니다. 
- 저런 디테일까지 신경쓸 필요는 없지만 우리가 전달하고 싶은 정보와 전체적인 구조에 집중할 수 있게 해줍니다! 

🧑‍💻 위의 그래프의 파라미터 설명 
- data : 불러오고자 하는 데이터 프레임
- kind : plot의 종류 
- x : x 축
- y : y 축
- col : 다른 facet으로 그래프를 subset할 변수
- hue : 변수의 그룹지정 -> 서로 다른 색상
- style : 변수의 그룸지정 -> 서로 다른 선 모양 대시선  
- size : 변수의 그룹지정 -> 서로 다른 사이즈 
- facet_kws=dict(sharex=False) : subplot의 x축을 공유하지 않기로 설정 

## Statistical estimation

- 가끔 우리는 평균값을 계산하는 함수를 다른 변수들에 적용하는 것에 관심이 있습니다.
- 많은 Seaborn 함수는 어떤 질문의 답을 위한 통계학적 예측을 자동으로 보여줍니다. 

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region", 
    hue="event", style="event");
```

🧑‍💻 그래프의 파라미터 설명 
- data : 불러오고자 하는 데이터 프레임
- kind : plot의 종류 
- x : x 축
- y : y 축
- col : 다른 facet으로 그래프를 subset할 변수
- hue : 변수의 그룹지정 -> 서로 다른 색상
- style : 변수의 그룸지정 -> 서로 다른 선 모양 대시선  

- 통계값이 측정됬을때 Seaborn은 자동적으로 신뢰구간을 계산하고 측정값의 불확실성은 error bar로 표시한다. 
- Seaborn에서의 통계학적 추정은 기술통계값보다 한 발 더 나아간다. 
- 예를 들어 lmplot() 함수를 통해 선형회귀모델과 그 모델의 불확실성을 scatter plot에 표현함으로써 그래프에 다양한 내용을 넣을 수 있다. 

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

- hue : 변수의 그룹지정 -> 서로 다른 색상

## Distributional representations 

- 데이터셋의 통계학적 분석은 변수 분포에 대한 지식을 필요로 합니다. 
- Seaborn 함수 displot() 분포를 시각화 하는데 몇 가지 접근법을 제공합니다. 
- 이 중에는 히스토그램과 커널 밀도 추정과 같은 계산 심화 계산법을 포함합니다. 

```python
sns.displot(data=tips, x='total_bill', col="time", kde=True)
```

- Seaborn은 익숙하지는 않지만 강력한 기술을 보여줍니다
- 경험적 누적분포의 그래프와 계산을 시각화 해줍니다. 

🧑‍💻 그래프의 파라미터 설명 
- data : 불러오고자 하는 데이터 프레임
- x : x 축
- col : 다른 facet으로 그래프를 subset할 변수
